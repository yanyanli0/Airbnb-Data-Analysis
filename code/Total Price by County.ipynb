```python
import pandas as pd
import numpy as np
import os
import glob
import altair as alt
import plotly.graph_objects as go
import plotly.express as px
```

```python
df=pd.read_csv("../data/processed/clean_airbnb.gz")
df.head()
```

# Transform (Monthly Price by County In NC)

```python
df=df[df['host_location'].str.split(', ').str[1] == 'NC']
```

```python
df.shape
```

```python
df['county'] = df['host_location'].str.split(', ').str[0]
df['county'].unique()
```

```python
df = df.set_index(pd.to_datetime(df['date']),drop=False)
df.head()
```

```python
monthly_mean = df.groupby([df.index.date, 'county'])['price'].agg(['mean', 'median','min','max'])
ummelted_df=monthly_mean.reset_index()
```

# Preview 

```python
ummelted_df=ummelted_df.rename(columns={"level_0":"date"})
ummelted_df
```

```python
alt.data_transformers.enable('json')
```

```python
selection = alt.selection_single(fields=['county'],name='Random')
color = alt.condition(selection,
                      alt.value('steelblue'),
                      alt.value('lightgray'))

bar=(alt.Chart(ummelted_df)
 .mark_bar()
 .encode(y='sum(mean):Q',
         x=alt.X('county:N',
         sort=alt.EncodingSortField(field='mean', op='mean', 
                            order='descending'),
         #tooltip=['mean(mean)','median(mean)','min(mean)','max(mean)']
         ),
         color=color
        )
 .interactive()
).add_selection(selection)


bar.title ="Total Revenue for Airbnb (NC State)"
bar.encoding.x.title = 'County'
bar.encoding.y.title = 'Total Revenue in Year 2023'



color2 = alt.condition(selection,
                      alt.Color('county:N',title="Total Revenue"),
#                       alt.value('steelblue'),
                      alt.value('lightgray'))

line1=(alt.Chart(ummelted_df)
 .mark_line(point=True)
 .encode(x=alt.X('month(date):T'),
         y='sum(mean):Q',
         color=color2
        )
)

line1.title ="Total Revenue for Airbnb (NC State)"
line1.encoding.x.title = 'Month'
line1.encoding.y.title = 'Aotal Revenue in Month(s)'
bar | line1
```
