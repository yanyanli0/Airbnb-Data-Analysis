```python
import pandas as pd
import zipfile

# North Carolina listing df
df=pd.read_csv("../data/raw/listings.csv.zip")
# North Carolina calendar df
ca=pd.read_csv("../data/raw/calendar.csv.zip")
```

```python
df.head()
```

```python
ca.head()
```

```python
merged_df=pd.merge(df, ca, left_on='id', right_on='listing_id', how='left')
merged_df.head()
```

```python
merged_df = merged_df.drop(columns=['listing_url','scrape_id','last_scraped', 'description','neighborhood_overview','picture_url','host_url','host_name',
                                    'host_about','host_is_superhost', 'host_thumbnail_url', 'host_picture_url', 'host_neighbourhood', 'host_listings_count', 
                                    'host_total_listings_count', 'host_verifications','host_has_profile_pic', 'host_identity_verified', 'neighbourhood',
                                    'neighbourhood_cleansed', 'neighbourhood_group_cleansed'])
```

```python
merged_df = merged_df.drop(columns=['listing_id', 'review_scores_accuracy', 'review_scores_cleanliness', 'review_scores_checkin', 'review_scores_communication',
                                    'review_scores_location', 'review_scores_value', 'license'])
```

```python
merged_df = merged_df.drop(columns=['minimum_minimum_nights', 'maximum_minimum_nights','minimum_maximum_nights', 'maximum_maximum_nights',
                                    'minimum_nights_avg_ntm', 'maximum_nights_avg_ntm', 'calendar_updated','has_availability', 'availability_30', 
                                    'availability_60','availability_90', 'availability_365', 'calendar_last_scraped', 'number_of_reviews_ltm', 
                                    'number_of_reviews_l30d','first_review', 'last_review', 'instant_bookable','minimum_nights_y', 'maximum_nights_y'])
```

```python
merged_df = merged_df.drop(columns=['price_y', 'adjusted_price', 'bathrooms_text', 'bathrooms'])
```

```python
merged_df.head()
```

```python
# rename
merged_df=merged_df.rename(columns={"price_x":"price",'minimum_nights_x':'minimum_nights', 'maximum_nights_x':'maximum_nights'})
merged_df.columns
```

```python
# remove $ in price column
merged_df['price'] = merged_df['price'].str.replace(',', '').str.replace('$', '')
merged_df['price'] = merged_df['price'].astype(float)
merged_df['price']
```

```python
# column types
merged_df.dtypes
```

```python
len(merged_df)
```

```python
# check if there are negative numbers
numeric_cols=merged_df.select_dtypes(include=['int', 'float']).columns
numeric_cols=numeric_cols.drop("longitude")
numeric_cols=numeric_cols.drop("latitude")
# numeric_cols=numeric_cols.drop("review_scores_rating")

# keep the rows have non negative numbers
df = merged_df.loc[(merged_df[numeric_cols] >= 0).all(axis=1)]
df.head()
```

```python
# rearrange the columns order
new_order=['id', 'name', 'price', 'date', 'available', 'room_type', 'accommodates', 'bedrooms', 'beds', 'amenities',  'latitude', 'longitude', 'property_type', 
           'host_id', 'host_since', 'host_location', 'host_response_time', 'host_response_rate', 'host_acceptance_rate', 
           'minimum_nights', 'maximum_nights', 'number_of_reviews', 'review_scores_rating', 'reviews_per_month', 
           'calculated_host_listings_count', 'calculated_host_listings_count_entire_homes', 
           'calculated_host_listings_count_private_rooms', 'calculated_host_listings_count_shared_rooms',]

df = df.reindex(columns=new_order)
df.head()
```

```python
# save to csv zipped file
df.to_csv('../data/processed/clean_airbnb.gz', compression='gzip', index=False)
```
